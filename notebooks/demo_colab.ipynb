## 1. Быстрая проверка: статический анализатор (работает без ключа)

```python
from reviewer import static_analyzer

sample_code = '''
def add_item(item, cart=[]):
    cart.append(item)
    return cart
'''

import json
print(json.dumps(static_analyzer.analyze(sample_code), ensure_ascii=False, indent=2))
```

## 2. Полный гибридный ревью (нужен GROQ_API_KEY)

```python
from reviewer import aggregator

mode = 'live' if os.environ.get('GROQ_API_KEY') else 'mock'
report = aggregator.run_review(sample_code, llm_mode=mode)
print('Режим:', mode)
print('Score:', report.score)
for f in report.findings:
    print(f['line'], f['source'], f['category'], '-', f['message'])
```
